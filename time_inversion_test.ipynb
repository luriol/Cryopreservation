```python
import numpy as np
import scipy.constants as scc
from matplotlib import pyplot as plt
# Start with function to give epsilon

# parameter values for 70% sucrose

# Ionization of water
# do quadratic fit to experimental data
T = np.array([0,10,20,30,40,50])
rho = np.array([85.841,43.297,23.844,14.082,8.810,5.791])*10000
a,b,c = np.polyfit(T,np.log(rho),2)
def rho_W(T):
    rho = np.exp(T**2*a+T*b+c)
    return rho 
def eps_sugar(omega,T,tau):
    Delta_k = 100
    y = -1j/(rho_W(T-273.15)*scc.centi)/omega/scc.epsilon_0/10
    y += Delta_k/(1 + 1j*omega*tau)
    return y

T = 260
omega = np.exp(np.linspace(np.log(1),np.log(1e8),1000))
tau = 1e-5
eps = eps_sugar(omega,T,tau)
plt.plot(omega/2/np.pi,np.log10(-np.imag(eps)))
plt.xscale("log")
plt.yscale("linear")
plt.ylim(-1,3.5)
plt.xlim(.01,1e6)
```
