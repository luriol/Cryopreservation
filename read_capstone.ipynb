```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_run_data(df, run_number):
    # Construct the label string based on the input number
    run_label = f"Run #{run_number}"

    # Step 1: Select columns with the matching run label
    run_columns = [col for col in df.columns if run_label in col]
    df_run = df[run_columns]

    # Step 2: Print the matching column names
    print(f"Columns for {run_label}:")
    for col in run_columns:
        print(col)

    # Step 3: Identify relevant channels
    time_col = next(col for col in run_columns if "Time" in col)
    ch_a_col = next(col for col in run_columns if "Ch A" in col)
    ch_v_col = next(col for col in run_columns if "Ch V" in col)

    # Step 4: Drop the last row of data
    time_data = df_run[time_col].iloc[:-1]
    ch_a_data = df_run[ch_a_col].iloc[:-1]
    ch_v_data = df_run[ch_v_col].iloc[:-1]

    # Step 5: Plot the signals
    plt.figure(figsize=(10, 6))
    plt.plot(time_data, ch_a_data, label="Ch A")
    plt.plot(time_data, ch_v_data, label="Ch V")
    plt.xlabel("Time")
    plt.ylabel("Signal")
    plt.title(f"{run_label}: Ch A and Ch V vs. Time")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage:
plot_run_data(df, 11)
```
